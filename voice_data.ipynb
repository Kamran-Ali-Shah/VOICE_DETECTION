```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from datasets import load_dataset
import time

# Initialize the model and pipeline
device = "cuda:0" if torch.cuda.is_available() else "cpu"
model_id = "openai/whisper-large-v3"
model = AutoModelForSpeechSeq2Seq.from_pretrained(model_id, use_safetensors=True)
model.to(device)
processor = AutoProcessor.from_pretrained(model_id)
pipe = pipeline(
"automatic-speech-recognition",
model=model,
tokenizer=processor.tokenizer,
feature_extractor=processor.feature_extractor,
max_new_tokens=128,
chunk_length_s=30,
batch_size=16,
return_timestamps=True,
device=device,
)

# Load the dataset
dataset = load_dataset("amansingh203/stuttering_asr")

# Measure the starting time
start_time = time.time()

# Transcribe every audio sample in the dataset
audio_count = 1
for split in dataset.keys():
    for example in dataset[split]:
        sample = example["audio"]
        # Measure the starting time for transcription
        transcription_start_time = time.time()
        result = pipe(sample)
        # Measure the ending time for transcription
        transcription_end_time = time.time()
        # Calculate the elapsed time for transcription
        transcription_elapsed_time = transcription_end_time - transcription_start_time
        print(f"Audio number: {audio_count}, Transcription elapsed time: {transcription_elapsed_time:.2f} seconds")
        audio_count += 1# Measure the ending time
end_time = time.time()

# Calculate and print the total elapsed time for all transcriptions
elapsed_time = end_time - start_time
print("Total elapsed time:", elapsed_time, "seconds")
```

```python
import pyarrow as pa
import pyarrow.parquet as pq
import pyarrow.feather as feather
from pydub import AudioSegment

def arrow_to_wav(arrow_file_path, wav_file_path):
  """
  Converts an Arrow file containing audio data to a WAV file.

  Args:
    arrow_file_path (str): Path to the Arrow file.
    wav_file_path (str): Path to save the WAV file.

  Raises:
    ValueError: If the Arrow file doesn't contain audio data.
  """

  # Check if it's a Parquet or Feather file
  if pq.ParquetDataset(arrow_file_path).schema:
    # Read the Parquet file using pyarrow
    table = pq.read_table(arrow_file_path)
  elif feather.FeatherDataset(arrow_file_path).schema:
    # Read the Feather file using pyarrow
    table = feather.read_table(arrow_file_path)
  else:
    raise ValueError("Unsupported Arrow file format. Please specify Parquet or Feather.")

  # Find the audio column by searching for data types like float32 or int16
  audio_column_name = None
  for col in table.schema.columns:
    if col.type in (pa.float32, pa.int16):
      audio_column_name = col.name
      break

  if audio_column_name is None:
    raise ValueError("Couldn't find audio data in the Arrow file.")

  # Extract the audio data as a NumPy array
  audio_data = table[audio_column_name].to_pandas().to_numpy()

  # Determine sample width and sample rate based on data type
  if audio_data.dtype == np.float32:
    sample_width = 4  # 32-bit float
  elif audio_data.dtype == np.int16:
    sample_width = 2  # 16-bit integer
  else:
    raise ValueError("Unsupported audio data type in the Arrow file.")

  # Extract sample rate from metadata (if available) or assume a common rate
  sample_rate = table.metadata.get("sample_rate", 16000)  # Adjust if needed

  # Create a pydub AudioSegment from the data
  audio_segment = AudioSegment(data=audio_data.tobytes(),
                              channels=1,
                              sample_width=sample_width,
                              frame_rate=sample_rate,
                              format="raw")

  # Export the AudioSegment to a WAV file
  audio_segment.export(wav_file_path, format="wav")

# Example usage:
arrow_file_path = "/content/drive/MyDrive/wav_data/train/data-00000-of-00003.arrow"
wav_file_path = "/content/drive/MyDrive/wav_data/train/data-00000-of-00003.wav"
arrow_to_wav(arrow_file_path, wav_file_path)

print(f"Converted Arrow file to WAV: {arrow_file_path} -> {wav_file_path}")
```
